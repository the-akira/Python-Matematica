# Onda Dente de Serra 

A [onda dente de serra](https://en.wikipedia.org/wiki/Sawtooth_wave) é um tipo de forma de onda não-senoidal. É assim denominada devido à sua semelhança com os dentes de uma serra de dentes planos com ângulo de saída zero.

Principais características da onda dente de serra:

- Uma onda dente de serra é uma forma de onda periódica e ela é não-sinusoidal.
- Uma onda dente de serra cresce para cima e cai bruscamente. 
- Uma onda dente de serra também pode ir para baixo e ascender bruscamente, neste caso ela é chamada de "onda dente de serra reversa" ou "onda dente de serra inversa".
- Ondas Dente de Serra tem suas aplicações em Sintetizadores Musicais, em Telas de Cathode-Ray Tube (CRT) e Osciloscópios.
- A função **signal.sawtooth** da biblioteca **SciPy** retorna uma forma de onda dente serra periódica ou uma forma de onda triangular.

Para vermos o exemplo, vamos então importar as bibliotecas necessárias:

```python
import matplotlib.pyplot as plt
from scipy import signal
import numpy as np
```

Criamos **1000** pontos separados linearmente com valores entre **0** e **1**:

```python
pontos_temporais = np.linspace(0, 1, 500)
```

Calculamos o sinal da onda dente de serra:

```python
sinal = signal.sawtooth(2 * np.pi * 5 * pontos_temporais)
```

Finalmente podemos plotar o gráfico:

```python
plt.figure(figsize=(16,9))
plt.plot(pontos_temporais, sinal, linewidth=3, color='blue')
plt.title('Onda Dente de Serra - 5 Hz amostrado em 1000Hz/segundo')
plt.xlabel('Tempo')
plt.ylabel('Amplitude')
plt.grid(True, which='both')
plt.axhline(y=0, color='k')
plt.ylim(-1.5, 1.5)
plt.show()
```
