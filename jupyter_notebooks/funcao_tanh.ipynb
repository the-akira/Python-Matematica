# Função Tangente Hiperbólica

**Tanh** é a função Tangente Hiperbólica, que é o análogo hiperbólico da função circular de Tangente usada em toda a trigonometria. Tanh é definido como a razão das funções seno hiperbólico e cosseno hiperbólico correspondentes. 

Na Matemática, as funções hiperbólicas são análogas às funções trigonométricas comuns.

As funções hiperbólicas básicas são:

- Seno hiperbólico "sinh"
- Cosseno hiperbólico "cosh"

No qual podemos derivar:

- Tangente hiperbólica "tanh"
- Cossecante hiperbólica "cosech"
- Secante hiperbólica "sech"
- Cotangente hiperbólica "coth"

A **tanh** ou **função de ativação tangente hiperbólica**, como também é conhecida, tem a seguinte fórmula matemática:

\begin{equation}
\tanh x=\frac{\cosh x}{\sinh x} = \frac{e^x - e^{-x}}{e^x + e^{-x}}
\end{equation}

Ela muito semelhante à função de ativação **sigmoid** e tem até a mesma forma de S.

Ela assume qualquer valor real como entrada e produz valores na faixa de $-1$ a $1$. Quanto maior a entrada (mais positiva), mais próximo o valor de saída estará de $1$, enquanto quanto menor a entrada (mais negativo), mais próximo a saída será $-1$.

Para vermos um exemplo da função Tanh, primeiro devemos importar as bibliotecas necessárias:

```python
import matplotlib.pyplot as plt
import numpy as np 
```

Definimos o nosso array de input **x**:

```python
x = np.linspace(-np.pi, np.pi, 50) 
```

Calculamos o nosso array de output **y**:

```python
y = np.tanh(x) 
```

Plotamos o gráfico da função Tanh:

```python
plt.figure(figsize=(16,9))
plt.plot(x, y, linewidth=3, color='b')
plt.axis([-3,3,-1,1])
plt.plot([-3,3],[0,0],'k--',linewidth=3)
plt.plot([0,0],[-3,3],'k--',linewidth=3)
plt.title("Função Tangente Hiperbólica") 
plt.xlabel("Eixo X") 
plt.ylabel("Eixo Y") 
plt.grid()
plt.show() 
```
