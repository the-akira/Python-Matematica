# Função de Unidade Linear Retificada

A função de Unidade Linear Retificada, conhecida por [ReLU](https://www.kaggle.com/dansbecker/rectified-linear-units-relu-in-deep-learning) (Rectified Linear Unit) é uma função de ativação que se tornou muito utilizada em modelos de Deep Learning. 

A função retorna $0$ se ela receber qualquer input negativo, porém para cada valor positivo de input $x$ ela retorna o valor de volta. 

Ela pode ser escrita da seguinte maneira:

\begin{equation}
f(x) = max(0,x)
\end{equation}

Para que possamos ver este exemplo, vamos primeiramente importar a biblioteca necessária:

```python
import matplotlib.pyplot as plt
```

Definimos a função **ReLU**:

```python
def rectified(x):
    return max(0.0, x)
```

O código a seguir gera uma série de inteiros entre **-10** até **10**:

```python
data_input = [x for x in range(-10, 11)]
print(data_input)
```

Calculamos os outputs para cada um de nossos inputs:

```python
data_output = [rectified(x) for x in data_input]
```

Finalmente podemos plotar o gráfico:

```python
plt.figure(figsize=(16,9))
plt.plot(data_input, data_output, linewidth=3, color='blue')
plt.grid()
plt.show()
```
