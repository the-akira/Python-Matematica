# Aplicando a Transformada de Fourier utilizando NumPy

- Transformada de Fourier é um dos conceitos mais aplicados no mundo da Ciência e Processamento de Sinal Digital
- Transformada de Fourier provê a representação do domínio de frequência do sinal original
- Por exemplo, dado um sinal sinusoidal no qual está no domínio de tempo, a Transformada de Fourier fornece as frequências de sinal constituintes.
- Utilizando a Transformada de Fourier, ambos sinais periódicos e não-periódicos podem ser transformados do domínio de tempo para o domínio de frequência

## Exemplo

O exemplo a seguir cria duas ondas senoidais e elas são adicionadas para criar um único sinal. Quando a Transformada de Fourier é aplicada para o sinal resultante, ela nos fornece os componentes de frequẽncia presentes na onda senoidal

```python
# Exemplo da Transformada de Fourier utilizando o método numpy.fft
import numpy as np
import matplotlib.pyplot as plotter
%matplotlib inline
plotter.rcParams["figure.figsize"] = [16,9]

# Quantos pontos de tempo são necessários, frequência de amostragem
samplingFrequency = 100;

# Em quais intervalos os pontos de tempo são amostradas
samplingInterval = 1 / samplingFrequency;

# Período de tempo inicial dos sinais
beginTime = 0;

# Período de tempo final dos sinais
endTime = 10; 

# Frequência dos sinais
signal1Frequency = 4;
signal2Frequency = 7;

# Pontos de tempo
time = np.arange(beginTime, endTime, samplingInterval);

# Cria duas ondas senoidais
amplitude1 = np.sin(2*np.pi*signal1Frequency*time)
amplitude2 = np.sin(2*np.pi*signal2Frequency*time)

# Cria um subplot
figure, axis = plotter.subplots(4, 1)
plotter.subplots_adjust(hspace=1)
 
# Representação do domínio de tempo para a onda senoidal 1
axis[0].set_title('Onda senoidal com frequência de 4 Hz')
axis[0].plot(time, amplitude1)
axis[0].set_xlabel('Tempo')
axis[0].set_ylabel('Amplitude')

# Representação do domínio de tempo para a onda senoidal 2
axis[1].set_title('Onda senoidal com frequência de 7 Hz')
axis[1].plot(time, amplitude2)
axis[1].set_xlabel('Tempo')
axis[1].set_ylabel('Amplitude')

# Adiciona as ondas senoidais
amplitude = amplitude1 + amplitude2

# # Representação do domínio de tempo para a onda senoidal resultante
axis[2].set_title('Onda senoidal com múltiplas frequências')
axis[2].plot(time, amplitude)
axis[2].set_xlabel('Tempo')
axis[2].set_ylabel('Amplitude')

# Representação do domínio de frequência
fourierTransform = np.fft.fft(amplitude)/len(amplitude)           # Amplitude normalizada
fourierTransform = fourierTransform[range(int(len(amplitude)/2))] # Exclui a frequência de amostragem

tpCount     = len(amplitude)
values      = np.arange(int(tpCount/2))
timePeriod  = tpCount/samplingFrequency
frequencies = values/timePeriod

# Representação do domínio de frequência
axis[3].set_title('Transformada de Fourier retratando os componentes de frequência')
axis[3].plot(frequencies, abs(fourierTransform))
axis[3].set_xlabel('Frequência')
axis[3].set_ylabel('Amplitude')

plotter.show()
```
